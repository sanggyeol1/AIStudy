```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/data/abalone.data')
df
```

```python
df.columns = ["Sex",	"Length",	"Diameter",	"Height",	"Whole weight",	"Shucked weight",	"Viscera weight",	"Shell weight",	"Rings"]
df
```

```python
from sklearn.preprocessing import StandardScaler, LabelEncoder # Sex를 숫자화

label = LabelEncoder()
df['Sex'] = label.fit_transform(df['Sex'])
df.head()
```

```python
x= df.drop('Rings',axis=1) #마지막 특성을 지워줘야 y값 대입 가능
x
```

```python
y = x['Diameter']
x= x.drop(columns=['Diameter'],axis=1)
y
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=423)
```

```python
from sklearn.metrics import mean_squared_error
from sklearn.metrics import confusion_matrix

from sklearn.tree import DecisionTreeRegressor

clf_dt = DecisionTreeRegressor()
clf_dt.fit(X_train, y_train)

pred_dt = clf_dt.predict(X_test)

print(pred_dt)
print(clf_dt.score(X_train, y_train))
```

```python
mse = np.sqrt(mean_squared_error(pred_dt, y_test))
print('평균제곱근오차', mse)
```

```python
from sklearn.ensemble import RandomForestRegressor

rf_clf = RandomForestRegressor()
rf_clf.fit(X_train, y_train)

pred_rf = rf_clf.predict(X_test)

print(pred_rf)
print(rf_clf.score(X_train, y_train))
```

```python
mse = np.sqrt(mean_squared_error(pred_rf, y_test))
print('평균제곱근오차', mse)
```

```python
from sklearn.linear_model import LinearRegression

clf_lr = LinearRegression()
clf_lr.fit(X_train, y_train)

pred_lr = clf_lr.predict(X_test)
print(pred_lr)
print(clf_lr.score(X_train, y_train))
```

```python
mse = np.sqrt(mean_squared_error(pred_lr, y_test))
print('평균제곱근오차', mse)
```

```python
from sklearn.svm import SVR

clf_svm = SVR()
clf_svm.fit(X_train, y_train)

pred_svm = clf_svm.predict(X_test)
print(pred_svm)
print(clf_svm.score(X_train, y_train))
```

```python
mse = np.sqrt(mean_squared_error(pred_svm, y_test))
print('평균제곱근오차', mse)
```
